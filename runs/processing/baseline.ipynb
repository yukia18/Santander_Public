# Formatting

```python
no_use_cols = [
    'target',
]
```

```python
use_cols = [col for col in train_df.columns if col not in no_use_cols]
```

```python
train_X = train_df.loc[:, use_cols + ['target']]
test_X = test_df.loc[:, use_cols]
```
